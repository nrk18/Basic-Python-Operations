### Requirement FR9 - Read CSV data from two files and merge it into a single Data Frame 

```python
# add code here

# Merging two files into a single Data Frame

import pandas as pd                                 # importing pandas library

# reading the files as a data frame using pandas
df1 = pd.read_csv("task2a.csv")                       
df2 = pd.read_csv("task2b.csv")

# Merging the two data frame by inner join and on the column common to both called region
df = df1.merge(df2, how = 'inner', on = 'region')

# Displaying the new Data Frame
df
```

### Requirement FR10 - Clean the merged data

```python
# add code here

# The merged data is cleaned 

import pandas as pd                                        # Pandas library imported

# Checking the condition whether a column exists before deleting
if 'share_voters_voted_to_the_right' in df.columns:
    del df["share_voters_voted_to_the_right"]

# Deleting all rows with NaN values
df = df.dropna(how='any',axis=0)  

# Resetting the index values after the deletion 
df = df.reset_index(drop=True)

# Displaying the result
df
```

### Requirement FR11 - Investigate the effects of household income on levels of bias crimes

```python
# add code here

# Code to check the effects of household income with levels of bias crimes

# importing libraries matplot lib and seaborn to create visualisation
import matplotlib.pyplot as plt
import seaborn as sns

# Customising the graph for better clarity
plt.figure(figsize=(14,8))

# The scatter plot to find the correlation is plotted 
sns.regplot(x=df["median_household_income"], y=df["avg_biascrimes_per_100k"])

# The axes and the graph is labelled
plt.xlabel("Household Income")
plt.ylabel("Levels of Bias Crimes")
plt.title('The effects of household income on levels of bias crimes')

# A brief of the findings from the graph is printed

print("The plot between household income and levels of bias crimes \
showed weak strength as the slope of the trend line was very low. The \
values also were also not falling near to the standard error. So, a \
linear correlation does not exist. ")
```

### Requirement FR12 - Test the hypothesis that household income has no effect on levels of bias crime

```python
# add code here

# The hypothesis is tested with the p values (Probability) using t-test
# The pearsonr function is imported from stats in scipy library
from scipy.stats import ttest_ind

# Hypothesis is tested between the variables household income and levels of bias crimes
result=ttest_ind(df["median_household_income"],df["avg_biascrimes_per_100k"]).pvalue

# The observation from the testing was printed as the result
print("The p value of household income vs levels of bias crime is :", result,".\
 Since the p value is less than 0.05, we reject the null hypothesis that\
household income has effect on levels of bias crime. So, we accept the \
alternate hypothesis that household income has no effect on levels of bias crime")
```

### Requirement FR14 - Test the hypothesis that urbanisation has a significant effect on levels of bias crimes

```python
# add code here

# The hypothesis is tested with the p values (Probability) using t-test
# The pearsonr function is imported from stats in scipy library
from scipy.stats import ttest_ind

# Hypothesis is tested between the variables household income and levels of bias crimes
result=ttest_ind(df["share_population_in_metro_areas"],df["avg_biascrimes_per_100k"]).pvalue

# The observation from the testing was printed as the result
print("The p value of urbanisation vs levels of bias crime is", result, ".\
Since the p value is less than 0.05, we reject the null hypothesis that \
urbanisation has effect on levels of bias crime. So, we accept the \
alternate hypothesis that urbanisation has no effect on levels of bias crime")
```
